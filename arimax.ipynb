```python
import os
import pandas as pd
import numpy as np

# --------------------------------------------------
# 1. LOAD ALL DATA
# --------------------------------------------------

def load_all_data(data_dir="data"):
    weather = pd.read_csv(os.path.join(data_dir, "formatted_weather.csv"))
    weather['Price Date'] = pd.to_datetime(
        weather['Price Date'], dayfirst=True, errors='coerce'
    )
    weather = weather.dropna(subset=['Price Date'])

    crop_data = {}
    for file in os.listdir(data_dir):
        if file.endswith(".csv") and file != "formatted_weather.csv":
            crop = file.replace(".csv", "")
            df = pd.read_csv(os.path.join(data_dir, file))
            df['Price Date'] = pd.to_datetime(
                df['Price Date'], dayfirst=True, errors='coerce'
            )
            df = df.dropna(subset=['Price Date'])
            crop_data[crop] = df

    print(f"Loaded {len(crop_data)} crop files.")
    return weather, crop_data


# --------------------------------------------------
# 2. ALIGN CROP & WEATHER (FINAL CORRECT LOGIC)
# --------------------------------------------------

def align_crop_weather(crop_df, weather_df):
    """
    Correct alignment:
    - Aggregate duplicate crop dates
    - DO NOT enforce daily frequency
    - Forward/backward fill weather only
    """

    # ---- Crop data ----
    crop = crop_df[['Price Date', 'Modal Price (Rs./Quintal)']].copy()

    # Aggregate duplicate dates
    crop = crop.groupby('Price Date', as_index=False).mean()

    # ---- Weather data ----
    weather = weather_df.copy()

    # ---- Merge on available crop dates ----
    merged = pd.merge(crop, weather, on='Price Date', how='left')
    merged = merged.sort_values('Price Date')

    # ---- Fill missing weather (key step) ----
    weather_cols = weather.columns.drop('Price Date')
    merged[weather_cols] = merged[weather_cols].ffill().bfill()

    # ---- Final safety cleaning ----
    merged = merged.replace([np.inf, -np.inf], np.nan)
    merged = merged.dropna()

    return merged


# --------------------------------------------------
# 3. TEST RUN
# --------------------------------------------------

if __name__ == "__main__":
    weather, crops = load_all_data()

    crop_name = list(crops.keys())[0]
    crop_df = crops[crop_name]

    aligned = align_crop_weather(crop_df, weather)

    print(f"\nAligned dataset for: {crop_name}")
    print(aligned.head())
    print("\nShape:", aligned.shape)
    print("Total NaNs:", aligned.isna().sum().sum())
```

```python
import os
import pandas as pd
import numpy as np

DATA_DIR = "data"

# --------------------------------------------------
# 1. LOAD DATA
# --------------------------------------------------

def load_all_data():
    weather = pd.read_csv(os.path.join(DATA_DIR, "formatted_weather.csv"))
    weather['Price Date'] = pd.to_datetime(
        weather['Price Date'], dayfirst=True, errors='coerce'
    )
    weather = weather.dropna(subset=['Price Date']).sort_values('Price Date')

    crops = {}
    for file in os.listdir(DATA_DIR):
        if file.endswith(".csv") and file != "formatted_weather.csv":
            name = file.replace(".csv", "")
            df = pd.read_csv(os.path.join(DATA_DIR, file))
            df['Price Date'] = pd.to_datetime(
                df['Price Date'], dayfirst=True, errors='coerce'
            )
            df = df.dropna(subset=['Price Date']).sort_values('Price Date')
            crops[name] = df

    print(f"Loaded {len(crops)} crop files.")
    return weather, crops


# --------------------------------------------------
# 2. ALIGN CROP & WEATHER (FINAL FIX)
# --------------------------------------------------

def align_crop_weather(crop_df, weather_df):
    """
    FINAL logic:
    - Restrict crop to weather period
    - Aggregate duplicate crop dates
    - Backward merge weather
    - DO NOT drop rows due to missing weather vars
    """

    # Restrict crop to weather date range
    start, end = weather_df['Price Date'].min(), weather_df['Price Date'].max()
    crop_df = crop_df[
        (crop_df['Price Date'] >= start) &
        (crop_df['Price Date'] <= end)
    ]

    # Crop price
    crop = crop_df[['Price Date', 'Modal Price (Rs./Quintal)']].copy()
    crop = crop.groupby('Price Date', as_index=False).mean()
    crop = crop.sort_values('Price Date')

    # Weather
    weather = weather_df.sort_values('Price Date')

    # Backward merge
    merged = pd.merge_asof(
        crop,
        weather,
        on='Price Date',
        direction='backward'
    )

    # Drop only rows where PRICE is missing
    merged = merged.dropna(subset=['Modal Price (Rs./Quintal)'])

    # OPTIONAL (clean): drop weather columns that are >90% NaN
    weather_cols = weather.columns.drop('Price Date')
    nan_ratio = merged[weather_cols].isna().mean()
    drop_cols = nan_ratio[nan_ratio > 0.9].index.tolist()
    merged = merged.drop(columns=drop_cols)

    return merged


# --------------------------------------------------
# 3. TEST RUN
# --------------------------------------------------

if __name__ == "__main__":
    weather, crops = load_all_data()

    crop_name = "apple"
    crop_df = crops[crop_name]

    aligned = align_crop_weather(crop_df, weather)

    print(f"\nAligned dataset for: {crop_name}")
    print(aligned.head())
    print("\nShape:", aligned.shape)
    print("NaNs per column:\n", aligned.isna().sum())
```

```python
import numpy as np
import pandas as pd

from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.preprocessing import MinMaxScaler

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping


# --------------------------------------------------
# ARIMAX + LSTM HYBRID MODEL
# --------------------------------------------------

def hybrid_arimax_lstm(
    aligned_df,
    forecast_horizon=90,
    lstm_window=30
):
    """
    aligned_df: output of align_crop_weather()
    forecast_horizon: number of days to forecast (90 = 3 months)
    """

    # -----------------------------
    # 1. Separate y and X
    # -----------------------------
    y = aligned_df['Modal Price (Rs./Quintal)']
    X = aligned_df.drop(
        columns=['Price Date', 'Modal Price (Rs./Quintal)'],
        errors='ignore'
    )

    # Handle missing exogenous values (VERY IMPORTANT)
    X = X.ffill().bfill()

    # -----------------------------
    # 2. ARIMAX MODEL
    # -----------------------------
    arimax = SARIMAX(
        y,
        exog=X,
        order=(2, 1, 2),
        enforce_stationarity=False,
        enforce_invertibility=False
    )

    arimax_fit = arimax.fit(disp=False)

    # Future exogenous values
    future_exog = pd.DataFrame(
        np.tile(X.iloc[-1].values, (forecast_horizon, 1)),
        columns=X.columns
    )

    arimax_forecast = arimax_fit.forecast(
        steps=forecast_horizon,
        exog=future_exog
    )

    # -----------------------------
    # 3. RESIDUALS FOR LSTM
    # -----------------------------
    residuals = y - arimax_fit.fittedvalues
    residuals = residuals.dropna()

    scaler = MinMaxScaler()
    scaled_res = scaler.fit_transform(residuals.values.reshape(-1, 1))

    X_lstm, y_lstm = [], []
    for i in range(lstm_window, len(scaled_res)):
        X_lstm.append(scaled_res[i - lstm_window:i])
        y_lstm.append(scaled_res[i])

    X_lstm = np.array(X_lstm)
    y_lstm = np.array(y_lstm)

    # -----------------------------
    # 4. LSTM MODEL
    # -----------------------------
    lstm = Sequential([
        LSTM(64, return_sequences=True, input_shape=(lstm_window, 1)),
        Dropout(0.2),
        LSTM(64),
        Dropout(0.2),
        Dense(1)
    ])

    lstm.compile(optimizer='adam', loss='mse')

    lstm.fit(
        X_lstm,
        y_lstm,
        epochs=50,
        batch_size=32,
        verbose=0,
        callbacks=[EarlyStopping(monitor='loss', patience=10)]
    )

    # -----------------------------
    # 5. LSTM RESIDUAL FORECAST
    # -----------------------------
    last_seq = scaled_res[-lstm_window:].reshape(1, lstm_window, 1)
    lstm_res_forecast = []

    for _ in range(forecast_horizon):
        pred = lstm.predict(last_seq, verbose=0)
        lstm_res_forecast.append(pred[0, 0])
        last_seq = np.roll(last_seq, -1, axis=1)
        last_seq[0, -1, 0] = pred[0, 0]

    lstm_res_forecast = scaler.inverse_transform(
        np.array(lstm_res_forecast).reshape(-1, 1)
    ).flatten()

    # -----------------------------
    # 6. FINAL HYBRID FORECAST
    # -----------------------------
    final_forecast = arimax_forecast.values + lstm_res_forecast

    return final_forecast
```
